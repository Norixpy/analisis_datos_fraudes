```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Convertir la fecha de nacimiento a edad
df["fecha_nacimiento"] = pd.to_datetime(df["fecha_nacimiento"], errors="coerce")
df["edad"] = (pd.to_datetime("today") - df["fecha_nacimiento"]).dt.days // 365  # Convertir dÃ­as a aÃ±os

# Separar los datos en dos grupos: Fraude vs No Fraude
edad_fraude = df[df["es_fraude"] == 1]["edad"].dropna()
edad_no_fraude = df[df["es_fraude"] == 0]["edad"].dropna()

# Prueba t de Student
t_stat, p_value = stats.ttest_ind(edad_fraude, edad_no_fraude, equal_var=False)

# InterpretaciÃ³n del resultado
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print(f"Rechazamos Hâ‚€: Hay una diferencia significativa en la edad entre fraudes y no fraudes (p = {p_value:.4f})")
else:
    print(f"No se rechaza Hâ‚€: No hay evidencia suficiente para afirmar que la edad afecta el fraude (p = {p_value:.4f})")

# VisualizaciÃ³n con histogramas y KDE plot
plt.figure(figsize=(12,6))
sns.kdeplot(edad_fraude, label="Fraude", shade=True, color="red")
sns.kdeplot(edad_no_fraude, label="No Fraude", shade=True, color="blue")
plt.title("DistribuciÃ³n de Edad en Transacciones Fraudulentas vs No Fraudulentas")
plt.xlabel("Edad")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear nuevas columnas basadas en fecha_hora_transaccion
df['hora'] = df['fecha_hora_transaccion'].dt.hour          # Hora de la transacciÃ³n
df['dia_semana'] = df['fecha_hora_transaccion'].dt.weekday  # DÃ­a de la semana (0=Lunes, 6=Domingo)
df['mes'] = df['fecha_hora_transaccion'].dt.month           # Mes de la transacciÃ³n
df['es_fin_de_semana'] = df['dia_semana'].apply(lambda x: 1 if x >= 5 else 0)  # Fin de semana

# OpciÃ³n 1: Graficar proporciones
def graficar_proporciones(df):
    # Calcular proporciones
    df_prop = df.groupby(['es_fin_de_semana', 'es_fraude']).size().reset_index(name='conteo')
    df_prop['proporcion'] = df_prop['conteo'] / df_prop.groupby('es_fin_de_semana')['conteo'].transform('sum')

    # Graficar proporciones
    plt.figure(figsize=(8, 6))
    sns.barplot(data=df_prop, x='es_fin_de_semana', y='proporcion', hue='es_fraude', palette='viridis')
    plt.title('ProporciÃ³n de Fraude en Fin de Semana vs DÃ­as Laborales', fontsize=14)
    plt.xlabel('Â¿Es Fin de Semana?', fontsize=12)
    plt.xticks([0, 1], labels=['DÃ­a Laboral', 'Fin de Semana'])
    plt.ylabel('ProporciÃ³n', fontsize=12)
    plt.legend(title='Es Fraude', labels=['No Fraude', 'Fraude'])
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# OpciÃ³n 2: Graficar con sobremuestreo
def graficar_sobremuestreo(df):
    # Sobremuestrear datos para balancear fraudes
    df_fraude = df[df['es_fraude'] == 1]
    df_no_fraude = df[df['es_fraude'] == 0].sample(n=len(df_fraude) * 3, random_state=42)  # Ratio 3:1
    df_balanceado = pd.concat([df_fraude, df_no_fraude])

    # Graficar
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df_balanceado, x='es_fin_de_semana', hue='es_fraude', palette='viridis')
    plt.title('Transacciones Balanceadas en Fin de Semana vs DÃ­as Laborales', fontsize=14)
    plt.xlabel('Â¿Es Fin de Semana?', fontsize=12)
    plt.xticks([0, 1], labels=['DÃ­a Laboral', 'Fin de Semana'])
    plt.ylabel('Cantidad de Transacciones', fontsize=12)
    plt.legend(title='Es Fraude', labels=['No Fraude', 'Fraude'])
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# OpciÃ³n 3: Graficar solo fraudes
def graficar_solo_fraudes(df):
    # Filtrar Ãºnicamente transacciones fraudulentas
    df_fraude = df[df['es_fraude'] == 1]

    # Graficar
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df_fraude, x='es_fin_de_semana', palette='Reds')
    plt.title('Transacciones Fraudulentas en Fin de Semana vs DÃ­as Laborales', fontsize=14)
    plt.xlabel('Â¿Es Fin de Semana?', fontsize=12)
    plt.xticks([0, 1], labels=['DÃ­a Laboral', 'Fin de Semana'])
    plt.ylabel('Cantidad de Transacciones Fraudulentas', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Llamar a las funciones
print("GrÃ¡fico de proporciones:")
graficar_proporciones(df)

print("GrÃ¡fico con sobremuestreo:")
graficar_sobremuestreo(df)

print("GrÃ¡fico solo de fraudes:")
graficar_solo_fraudes(df)

```

objective: 'binary:logistic'

Define la funciÃ³n objetivo que el modelo debe optimizar.

'binary:logistic' significa que el modelo realizarÃ¡ clasificaciÃ³n binaria y producirÃ¡ probabilidades como salida (valores entre 0 y 1).

eval_metric: 'auc'

Es la mÃ©trica de evaluaciÃ³n utilizada para medir el rendimiento del modelo en cada iteraciÃ³n.

'auc' (Ãrea Bajo la Curva ROC) es una mÃ©trica comÃºn para problemas de clasificaciÃ³n binaria que mide quÃ© tan bien el modelo distingue entre clases.

max_depth: 6

Indica la profundidad mÃ¡xima de los Ã¡rboles de decisiÃ³n en el modelo.

Un valor mayor puede capturar mÃ¡s relaciones complejas en los datos, pero tambiÃ©n aumenta el riesgo de sobreajuste.

eta: 0.3

Es la tasa de aprendizaje (learning rate), que controla cuÃ¡nto cambian los pesos en cada iteraciÃ³n.

Un valor mÃ¡s alto acelera el entrenamiento pero puede hacer que el modelo converja a una soluciÃ³n subÃ³ptima.

Un valor mÃ¡s bajo hace que el modelo aprenda mÃ¡s lentamente, pero puede mejorar la generalizaciÃ³n.

seed: 42

Es la semilla aleatoria para garantizar la reproducibilidad de los resultados.

Usar la misma semilla permite obtener los mismos resultados al entrenar el modelo varias veces.
